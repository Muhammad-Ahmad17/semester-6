# Lab 6 Manual Guide: Comparison of Different Regression Algorithms

## Objective

This lab compares three regression models on the same dataset:
- Linear Regression
- Decision Tree Regression
- Random Forest Regression

The goal is to train each model, evaluate it with mean absolute error (MAE), and compare the results on both the training set and the test set.

## Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
```

### importing the libraries

`pandas` is used for loading the dataset, `numpy` supports numerical operations, and scikit-learn provides the regression models and the MAE metric used for evaluation.

## Importing the dataset

```python
df = pd.read_csv('50_Startups.csv')
df.head()
```

### loading the dataset

The `50_Startups.csv` file contains startup data used to predict profit. The notebook first loads the CSV file and previews the first few rows so the structure of the dataset can be inspected before modeling.

# Splitting the data set into features(Independent Variables) and Outcome(dependent variable)

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train ,y_test = train_test_split (df[['R&D Spend', 'Administration', 'Marketing Spend']],df['Profit'],test_size=0.33,random_state=3)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

### splitting the data set into features and target

`X` stores the independent variables and `y` stores the dependent variable. The dataset is then split into training and test sets so each model can be trained on one part of the data and checked on unseen data for fair comparison.

# **Linear Regression**

### linear regression

This section fits a linear regression model on the training data and prints the MAE for both the training and test sets. The lower the MAE, the better the model fits the data.

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)

# Print coefficients
print(f"Intercept (b0) = {model1.intercept_:.2f}")
print(f"Slope (b1) = {model1.coef_[0]:.2f}")
print(f"Equation: y = {model1.intercept_:.2f} + {model1.coef_[0]:.2f}x")
print()

print("Linear Regression")
print("====================")
y_pred_train1 = model1.predict(X_train)
MAE_train = mean_absolute_error(y_train, y_pred_train1)
print("Linear Regression Train set: MAE ${:.2f}".format(MAE_train))

y_pred_test1 = model1.predict(X_test)
MAE_test = mean_absolute_error(y_test, y_pred_test1)
print("Linear Regression Test set: MAE ${:.2f}".format(MAE_test))
print("====================\n")


```

# **Decision Tree Regression**

### decision tree regression

The decision tree model learns a set of split rules from the training data. Here, `max_depth=2` keeps the tree small and easier to compare with the other regression models.

```python

model2 = tree.DecisionTreeRegressor(max_depth=2, random_state = 3)
model2.fit(X_train, y_train)

print("Decision Tree")
print("==============================")
y_pred_train2 = model2.predict(X_train)
MAE_train2 = mean_absolute_error(y_train, y_pred_train2)
print("Decision Tree Train set: MAE ${:.2f}".format(MAE_train2))

y_pred_test2 = model2.predict(X_test)
MAE_test2 = mean_absolute_error(y_test, y_pred_test2)
print("Decision Tree Test set: MAE ${:.2f}".format(MAE_test2))
print("==============================\n")
```

# **Random forest Regression**

### random forest regression

Random forest combines many decision trees to reduce overfitting and usually gives a stronger baseline than a single tree. This section trains the model and reports MAE on the training and test sets.

```python

model3 = RandomForestRegressor(random_state=3)
model3.fit(X_train, y_train)

print("Random Forest Regressor")
print("==================================================")
y_pred_train3 = model3.predict(X_train)
MAE_train3 = mean_absolute_error(y_train, y_pred_train3)
print("Random Forest Regressor TrainSet: MAE ${:.2f}".format(MAE_train3))

y_pred_test3 = model3.predict(X_test)
MAE_test3 = mean_absolute_error(y_test, y_pred_test3)
print("Random Forest Regressor TestSet: MAE ${:.2f}".format(MAE_test3))
print("==================================================\n")
```

## Conclusion

After running the notebook, compare the train and test MAE values for all three models. The preferred model is the one with the lowest test MAE, because it generalizes better to unseen data.
