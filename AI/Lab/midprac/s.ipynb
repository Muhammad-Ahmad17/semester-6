```python
import pandas as pd
from sklearn.model_selection import train_test_split

dataset = pd.read_csv('./Salary_Data.csv')
```

```python
dataset.head()
```

```python

x_train , x_test , y_train ,y_test = train_test_split (dataset[['YearsExperience']],dataset['Salary'],test_size=0.33,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
fitting= model.fit(x_train,y_train)
y_hat = model.predict(x_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test,y_test,color='red')
plt.scatter(x_test,y_hat)
plt.legend(['orignal','test'])
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mean_err = r2_score(y_test,y_hat)
mean_err
```

```python
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(random_state=0 , max_depth= 3 , min_samples_leaf= 2 ,min_samples_split= 5)
```

```python
model.fit(x_train,y_train)
y_hat= model.predict(x_test)
```

```python
plot_tree(model)
```

```python
r2=r2_score(y_test,y_hat)
r2
```
