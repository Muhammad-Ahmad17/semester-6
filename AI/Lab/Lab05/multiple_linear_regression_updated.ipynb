# Multiple Linear Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
```

## Importing the dataset

```
df = pd.read_csv('50_Startups.csv')
df.head()
```

```
# Correct way to filter for numeric columns
df = df.select_dtypes(include=['number'])
```

```
df.head()
```

```
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```


#Correlation Matrix
cor = df.corr()
print(cor)

```

```
print(X)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Multiple Linear Regression model on the Training set

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```
# Calculate and print RMSE for training set
print("Regression")
print("====================")
y_pred_train = regressor.predict(X_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
print("Regression Train set: RMSE {}".format(RMSE_train))

# Predict and evaluate on the testing set
y_pred = regressor.predict(X_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, y_pred))
print("====================")
print("Regression Test set: RMSE {}".format(RMSE_test))


```

```
# Code for obtainging Model coefficients along with their relevant categories
feature_names = df.columns[:-1] # Gets R&D Spend, Administration, Marketing Spend
coef_dict = {}

for coef, feat in zip(regressor.coef_, feature_names):
    coef_dict[feat] = float(coef)  # <--- Cast to standard Python float here

print("Model Coefficients:")
print(coef_dict)
```

```
mrd=regressor.coef_[0] #b1 which is Model Coefficient of R&D Spend
```

```
mad=regressor.coef_[1] #b2 which is Model Coefficient of Administration
```

```
mms=regressor.coef_[2]  #b3 which is Model Coefficient of Marketing Spend
```

```
c=regressor.intercept_ # b0 which is intercept
```

```
X_test[0] # first sample of test set
```

```
float(X_test[0][0]*mrd + X_test[0][1]*mad + X_test[0][2]*mms + c)
```
