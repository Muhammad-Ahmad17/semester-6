```python
def count_even_number (number_list):
    even_count = 0
    for num in number_list:
        if num % 2 == 0:
            even_count += 1
    return even_count
```

### inlab task 2

```python
def get_grade_data(marks):
    """Returns (Grade, Grade Points) based on lab table """
    if marks >= 85: return "A", 4.00
    elif marks >= 80: return "A-", 3.66
    elif marks >= 75: return "B+", 3.33
    elif marks >= 71: return "B", 3.00
    elif marks >= 68: return "B-", 2.66
    elif marks >= 64: return "C+", 2.33
    elif marks >= 61: return "C", 2.00
    elif marks >= 58: return "C-", 1.66
    elif marks >= 54: return "D+", 1.30
    elif marks >= 50: return "D", 1.00
    else: return "F", 0.00
```

```python

def calculate_gpa(student_list):
    results = []
    for student in student_list:
        course_grades = []
        points = []
        for mark in student['marks']:
            grade, pt = get_grade_data(mark)
            course_grades.append(grade)
            points.append(pt)

        # GPA is the average of Grade Points
        avg_gpa = sum(points) / len(points)

        results.append({
            'name': student['name'],
            'grades': course_grades,
            'grade_points': points,
            'gpa': round(avg_gpa, 2)
        })
    return results
```

```python
# Student records for testing
students = [
    {'name': 'Student 1', 'marks': [88, 72, 65, 90, 55]},
    {'name': 'Student 2', 'marks': [45, 50, 60, 58, 62]},
    {'name': 'Student 3', 'marks': [80, 82, 84, 78, 81]},
    {'name': 'Student 4', 'marks': [75, 76, 74, 70, 68]},
    {'name': 'Student 5', 'marks': [95, 92, 89, 88, 91]}
]

print(calculate_gpa(students))
```
