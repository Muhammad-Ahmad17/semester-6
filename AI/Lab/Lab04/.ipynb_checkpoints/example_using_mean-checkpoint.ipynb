# Data Preprocessing with Outlier Detection
Based on Lab 4  and data.csv

```
# 1. Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```
# 2. Importing the dataset
# Ensure your 'data.csv' file is in the same directory as this notebook
dataset = pd.read_csv('data.csv')
dataset.head()
```

```
dataset.describe()
```

```
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
# 3. Visualizing Outliers with Box Plots
# We use box plots to visualize the distribution and identify outliers in numerical columns.

plt.figure(figsize=(12, 5))

# Box plot for Age
plt.subplot(1, 2, 1)
sns.boxplot(y=dataset['Age'], color='skyblue')
plt.title('Box Plot of Age')

# Box plot for Salary
plt.subplot(1, 2, 2)
sns.boxplot(y=dataset['Salary'], color='green')
plt.title('Box Plot of Salary')

plt.show()
```

```

# 4. Handling Outliers (IQR Method)
# This method identifies outliers using the Interquartile Range (IQR).
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers detected in {column}:")
    print(outliers if not outliers.empty else "No outliers found.")
    return lower_bound, upper_bound
```

```
age_low, age_high = detect_outliers_iqr(dataset, 'Age')
sal_low, sal_high = detect_outliers_iqr(dataset, 'Salary')
```

```
# 5. Taking care of missing data
# We use the mean strategy for the numerical columns: Age (index 1) and Salary (index 2)
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```
# 6. Encoding categorical data
# Independent Variable (Country) using OneHot Encoding
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

# Dependent Variable (Purchased) using Label Encoding
le = LabelEncoder()
y = le.fit_transform(y)
```

```

# 7. Splitting the dataset into Training and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```


# 8. Feature Scaling
# We scale columns starting from index 4 to avoid scaling the dummy variables created by OneHotEncoder.
sc = StandardScaler()
X_train[:, 4:] = sc.fit_transform(X_train[:, 4:])
X_test[:, 4:] = sc.transform(X_test[:, 4:])

# Displaying results
print("\n--- Training Set Scaled version ---")
print(X_train)
print("\n--- Test Set Scaled version ---")
print(X_test)
```
