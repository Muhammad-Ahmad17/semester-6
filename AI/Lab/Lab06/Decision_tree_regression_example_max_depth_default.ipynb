## Importing the libraries

```
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn import tree
import matplotlib.pyplot as plt
```

## Importing the dataset

```
df = pd.read_csv('50_Startups.csv')
df.head()
```

```
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

# Predicting the results for Decision Tree Regression

```

model2 = tree.DecisionTreeRegressor(random_state = 3)
model2.fit(X_train, y_train)

print("Decision Tree")
print("==============================")

# Training set evaluation
y_pred_train2 = model2.predict(X_train)
# Wrapping MSE in np.sqrt() to get the actual Root Mean Squared Error
RMSE_train2 = np.sqrt(mean_squared_error(y_train, y_pred_train2))
print("Decision Tree Train set: RMSE {}".format(RMSE_train2))

# Test set evaluation
y_pred_test2 = model2.predict(X_test)
RMSE_test2 = np.sqrt(mean_squared_error(y_test, y_pred_test2))
print("Decision Tree Test set: RMSE {}".format(RMSE_test2))

print("==============================")
```

```


# 1. Visualising the Decision Tree Regressor
plt.figure(figsize=(20, 10))
tree.plot_tree(model2,
               feature_names=df.columns[:-1],
               filled=True,
               rounded=True,
               fontsize=10)
plt.title("Decision Tree Visualization")
plt.show()
```
