```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# 1. LOAD THE NUMERICAL DATASET
# This file contains:  Age, Tenure, Balance, EstimatedSalary, Exited
dataset = pd.read_excel('Churn_Modelling.xlsx')
X = dataset.iloc[:, :-1].values  # All features (columns 0 to 3)
y = dataset.iloc[:, -1].values   # The target (Exited)
```

```python

# 2. SPLIT INTO TRAINING AND TEST SET
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 3. FEATURE SCALING (Crucial for Neural Networks)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#4. Build the ANN model
model = Sequential()

# Step 1: Add the input layer and the first hidden layer with ReLU activation
# (Uses 6 neurons)
model.add(Dense(units=6, activation='relu'))

# Step 2: Add the second hidden layer with ReLU activation
model.add(Dense(units=6, activation='relu'))

# Step 3: Add the output layer with Sigmoid activation
# (Relevant for binary classification in the ANN example)
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
# Using 'binary_crossentropy' for the ANN's binary classification task
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=32, epochs=150)

```

```python
# 5. EVALUATE THE MODEL
y_pred = (model.predict(X_test) > 0.5)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
