### reading dataset

```python
import pandas as pd 
df = pd.read_csv('./USA_Housing.csv')
```

### checking data set

```python
df.head()
```

## extracting features and outcomes from the dataset 
### x -> features
### y -> outcome

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms','Area Population']]
```

```python
y = df['Price']
```

## splitting dataset into training and testing , with ratio of 80 and 20

```python
xtrain, xtest ,ytrain ,ytest = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=3
    )
```

# Multiple Linear Regression


```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain) # fitting data into model
```

```python
y_hat_test_mlr = model.predict(xtest) # prediction on test data
y_hat_test_mlr # prediction result
```

### coefcent and intercept of multiple linear regression

```python
intercept = model.intercept_
print ("intercept of our model is :" , intercept)
coefcent = model.coef_
print ('coefcent for Avg. Area Income', coefcent[0])
print ('coefcent for Avg. Area House Age', coefcent[1])
print ('coefcent for Avg. Area Number of Rooms', coefcent[2])
print ('coefcent for Avg. Area Number of Bedrooms', coefcent[3])
print ('coefcent for Area Population', coefcent[4])
```

## model evaluation and comparison for MLR  

```python
from sklearn.metrics import mean_absolute_error ,r2_score
```

```python
mae_test_mlr = mean_absolute_error (ytest,y_hat_test_mlr)
print ('mean square error of MLR:', mae_test_mlr)
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor()
```

```python
model1.fit(xtrain,ytrain)
```

```python
y_hat_test_dtr = model1.predict(xtest) # prediction on test data
y_hat_test_dtr # prediction result
```

## model evaluation and comparison for DTR 

```python
mae_test_dtr = mean_absolute_error (ytest,y_hat_test_dtr)
print ('mean square error of DTR:', mae_test_dtr)
```

# comparing performance

```python
print ('mean square error of MLR:', mae_test_mlr)
print ('mean square error of DTR:', mae_test_dtr)
```

#### MAE of Multiple linear regression < MAE of Decision Tree regression
#### Multiple linear regression is more accurate then Decision Tree regression

### also verifing through r2 score  , r2 score = 1 for ideal case , r2 near to 1 is more accurate 

```python
r2_mlr = r2_score( ytest, y_hat_test_mlr)
r2_dtr = r2_score(ytest,y_hat_test_dtr)

r2_dtr , r2_mlr
```

#### R2-score of Multiple linear regression < R2-score of Decision Tree regression
