```python
plt.figure(figsize=(15, 8))
df_without_price = df.drop('Price', axis=1)
sns.boxplot(df_without_price)
plt.grid()
plt.show()
```

```python
# 1. Calculate the 99th percentile for every column
quantiles = df.quantile(0.99)

# 2. Filter rows where EVERY column is below its respective 99th percentile
df1 = df[(df < quantiles).all(axis=1)]

# Now you can proceed with your plotting
df1_without_price = df1.drop('Price', axis=1)
sns.boxplot(df1_without_price)
plt.show()
```

```python
plt.figure(figsize=(15, 8))

# 1. Calculate the 1st and 99th percentiles for all numeric columns at once
quantile1 = df.quantile(0.01)
quantile99 = df.quantile(0.99)

# 2. Create a boolean mask: True if a value is within the range [Q1, Q99]
# .all(axis=1) ensures we only keep rows where EVERY column satisfies the condition
condition = ((df > quantile1) & (df < quantile99)).all(axis=1)
df_filtered = df[condition]

# 3. Drop 'Price' for visualization purposes as before
df_filtered_without_price = df_filtered.drop('Price', axis=1)

# 4. Plot the results
sns.boxplot(data=df_filtered_without_price)
plt.grid()
plt.show()
```
