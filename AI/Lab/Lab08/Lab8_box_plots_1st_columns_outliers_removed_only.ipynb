```python
# Importing libraries needed


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



```

```python
df = pd.read_csv("Housing.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(18, 8))
sns.boxplot(df)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
boxplot = pd.DataFrame(df).boxplot()
```

```python
#boxplot = pd.DataFrame(df).boxplot()
plt.figure(figsize=(15, 8))
boxplot = df['Avg. Area Income'].plot.box()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(df['Avg. Area Income'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
df_without_price = df.drop('Price', axis=1)
sns.boxplot(df_without_price)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
quantile99 = df.iloc[:,0].quantile(0.99)
df1 = df[df.iloc[:,0] < quantile99]
df1_without_price = df1.drop('Price', axis=1)
sns.boxplot(df1_without_price)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
quantile1 = df.iloc[:,0].quantile(0.01)
quantile99 = df.iloc[:,0].quantile(0.99)
df2 = df[(df.iloc[:,0] > quantile1) & (df.iloc[:,0] < quantile99)]
df2_without_price = df2.drop('Price', axis=1)
sns.boxplot(data=df2_without_price)
plt.grid()
plt.show()
```
